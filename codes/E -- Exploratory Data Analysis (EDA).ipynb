## Background

Deep fakes (a combination of "[*deep learning*](https://en.wikipedia.org/wiki/Deep_learning)" and "fake") are synthetic media in which a person in an existing image or video is replaced with someone similar. While the creation of fake content is not new, deep fakes leverage powerful techniques from artificial intelligence and machine learning to produce audio and/or visual content that can more easily deceive.

The main machine learning methods used to create deep fakes are based on deep learning and involve training generative neural network architectures, such as [*generative adversarial networks (GANs)*](https://en.wikipedia.org/wiki/Generative_adversarial_network).

Deep fakes have many potential applications across multiple industries and domains, including but not limited to:
 - [*Art*](https://digitalsynopsis.com/design/samsung-living-portrait/)
 - [*Charity*](https://abcnews.go.com/International/david-beckham-speaks-languages-campaign-end-malaria/story?id=62270227)
 - [*Entertainment*](https://youtu.be/mPU0WNUzsBo)
 - [*Gaming*](https://www.theverge.com/2018/12/3/18121198/ai-generated-video-game-graphics-nvidia-driving-demo-neurips)
 - [*Healthcare*](https://www.fastcompany.com/90240746/deepfakes-for-good-why-researchers-are-using-ai-for-synthetic-health-data)
 - [*History*](https://www.theverge.com/2019/5/10/18540953/salvador-dali-lives-deepfake-museum)
 - [*Public Service*](https://youtu.be/cQ54GDm1eL0)
 - etc.

However, deep fakes garnered widespread attention for their uses in creating child sexual abuse material, [*celebrity pornographic videos*](https://www.highsnobiety.com/p/what-are-deepfakes-ai-porn/), [*revenge porn*](https://en.wikipedia.org/wiki/Revenge_porn), [*fake news*](https://en.wikipedia.org/wiki/Fake_news), hoaxes, bullying, financial fraud, etc. This has elicited responses from both industry ([*Source*](https://thenextweb.com/news/twitter-pornhub-and-other-platforms-ban-ai-generated-celebrity-porn)) and government ([*Source*](https://www.congress.gov/bill/116th-congress/house-bill/3230)) to detect and limit their use. ## Trusted Media Challenge (TMC) Dataset

The dataset used here is the Trusted Media Challenge (TMC) dataset released by [*AI Singapore*](https://aisingapore.org/), comprising of 4,380 fake and 2,563 real videos, with various video and/or audio manipulation methods employed to produce different types of fake media.

**Key Features of the TMC dataset**:
1. Videos in dataset focus on Asian content and ethnicities.
2. Videos (with audios included) vary in durations, background, illumination, and may contain perturbations that mimic transmission errors and compression.
3. The 222 consented subjects in these videos are either Professionals or Amateurs:
   - **Professionals**: [*CNA*](https://www.channelnewsasia.com/) news presenters and journalists talking in news programs, and [*ST*](https://www.straitstimes.com/) interviewees answering questions in interview programs.
   - **Amateurs**: Freelancers talking about different topics like interest, movies, food, etc.
4. There are 4 different types of fake media in the TMC dataset:

| **Type**   | **Video**  | **Audio**  | **Percentage** |
|------------|------------|------------|----------------|
| 1          | Fake       | Fake       | 20.24%         |
| 2          | Fake       | Real       | 22.97%         |
| 3          | Real       | Fake       | 9.07%          |
| 4          | Real       | Real       | 10.80%         |
| **_[ Blank ]_** | **_Real_** | **_Real_** | **_36.92%_**   |

5. For Type-4 (See above), while both the videos and audios are real, the speech content does not match the mouth movement in the videos i.e. the video footage from one real video is combined with the audio segment from another real video to create a new medium. Therefore, it is regarded as fake medium too.

Refer to the **[*AI Singapore Trusted Media Challenge Dataset*](https://arxiv.org/abs/2201.04788.pdf)** paper submitted by [*AI Singapore*](https://aisingapore.org/) dated 08 Mar 2022 for details.

The TMC dataset is also available for research purpose upon request via [*tmc-dataset@aisingapore.org*](mailto:tmc-dataset@aisingapore.org).

### Display 3 samples of fake & real images each

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
%matplotlib inline

# Directories where training video & image files are stored
train_dir = '.\\split_dataset\\train\\'

# Get list of pngs files from image class sub folders
real_imgs = [fn for fn in os.listdir(f'{train_dir}0') if fn.endswith('.png')]
fake_imgs = [fn for fn in os.listdir(f'{train_dir}1') if fn.endswith('.png')]

# Randomly select 3 of each
select_real = np.random.choice(real_imgs, 3, replace = False)
select_fake = np.random.choice(fake_imgs, 3, replace = False)

# Plot 2 x 3 image matrix
fig = plt.figure(figsize = (8, 6))
for i in range(6):
    if i < 3:
        fp = f'{train_dir}/0/{select_real[i]}'
        label = 'REAL'
    else:
        fp = f'{train_dir}/1/{select_fake[i - 3]}'
        label = 'FAKE'
    ax = fig.add_subplot(2, 3, i + 1)
    
    # Plot in reversed version of grayscale (i.e. cmap = 'Greys_r') without rescaling, remove target_size
    # Otherwise, images appear like film negatives when plotted in Grayscale (i.e. cmap = 'Greys')
    fn = image.load_img(fp, target_size = (100, 100), color_mode = 'grayscale')    # Grayscale
    plt.imshow(fn, cmap = 'Greys_r')    # Reversed version of Grayscale
    plt.title(label, fontweight = 'bold')
    plt.axis('off')
plt.show()

# Check total number of real and fake images
len(real_imgs), len(fake_imgs)
```

Observation(s):
- Watermark appears near the bottom right of many cropped faces which may affect the accuracy of the CNN model.
- Some of the 12 video perturbations included in the TMC dataset may be observed on some of these samples (e.g. bottom left and bottom right images).
- Faces are mainly of Asian ethnicities and includes female subjects.

### Create matrix from all images & convert into a vector or 1D array

```python
# Create n X m matrix
def img2np(path, list_of_filename, size = (256, 256)):
    # Iterate through each file
    for fn in list_of_filename:
        fp = path + fn
        current_image = image.load_img(fp, target_size = size, color_mode = 'grayscale')    # 0 = Black, 255 = White
        # Convert image into matrix
        img_ts = image.img_to_array(current_image)
        # Turn matrix into a vector / 1D array
        img_ts = [img_ts.ravel()]
        try:
            # Concatenate the different images
            full_mat = np.concatenate((full_mat, img_ts))
        except UnboundLocalError: 
            # If not assigned yet, assign one
            full_mat = img_ts
    return full_mat

# Run it on the 2 subfolders
real_images = img2np(f'{train_dir}/0/', real_imgs)
fake_images = img2np(f'{train_dir}/1/', fake_imgs)
```

### Plot Mean Real Image & Mean Fake Image

```python
def find_mean_img(full_mat, title, size = (256, 256)):
    # Calculate the average or mean image
    mean_img = np.mean(full_mat, axis = 0)
    # Reshape the mean image back into a matrix
    mean_img = mean_img.reshape(size)
    plt.imshow(mean_img, vmin = 0, vmax = 255, cmap = 'Greys_r')    # 0 = White, 255 = Black 
    plt.title(f'Average {title}', fontweight = 'bold')
    plt.axis('off')
    plt.show()
    return mean_img

# Print both the mean real image and the mean fake image
real_mean = find_mean_img(real_images, 'REAL')
fake_mean = find_mean_img(fake_images, 'FAKE')
```

Observation(s):
- Both images look quite similar at first glance.
- On closer inspection, the forehead, nosebridge, chin and both cheek regions below the eyes appear brighter (i.e. lighter shade of grey) in the Average REAL image versus the Average FAKE image.

### Plot contrast between Mean Images

```python
# Contrast Between Average Images
contrast_mean = real_mean - fake_mean

# 0 = blue (No Observable difference), 
# 255 = red (Observable difference, potentially differentiate real from fake)
plt.imshow(contrast_mean, cmap = 'bwr')
plt.title(f'Difference Between Real & Fake Average', fontweight = 'bold')
plt.axis('off')
plt.show()
```

Observation(s):
- The face and upper body regions appear red which is reasonably expected since any deepfake video manipulations made would generally focus on these regions too. 
- Reference to the 6 sample images above, a portion of the watermark appears near the bottom right of 5 of the 6 sample cropped faces. This potentially explains why the bottom right and right centre regions appear red in this contrast average image.

### Plot Standard Deviation Real Image & Standard Deviation Fake Image

```python
def find_std_img(full_mat, title, size = (256, 256)):
    # calculate the standard deviation
    std_img = np.std(full_mat, axis = 0)
    # reshape it back to a matrix
    std_img = std_img.reshape(size)
    plt.imshow(std_img, vmin = 0, vmax = 255, cmap = 'Greys_r')    # 0 = White, 255 = Black
    plt.title(f'Standard Deviation {title}', fontweight = 'bold')
    plt.axis('off')
    plt.show()
    return std_img

real_std = find_std_img(real_images, 'REAL')
fake_std = find_std_img(fake_images, 'FAKE')
```

Observation(s):
- Similar to the pair of Average images above, both images look quite alike at first glance.
- On closer inspection, the regions around key facial landmarks such as the eyebrows, the chin, bottom portion of the nose and below both eyes appear blurrer or less defined in the Standard Deviation FAKE image versus the Standard Deviation REAL image.

### Plot contrast between Standard Deviation Images

```python
# Contrast Between Standard Deviation Images
contrast_std = real_std - fake_std

# 0 = blue (No Observable difference), 
# 255 = red (Observable difference, potentially differentiate real from fake)
plt.imshow(contrast_std, cmap = 'bwr')
plt.title(f'Difference Between Real & Fake Standard Deviation', fontweight = 'bold')
plt.axis('off')
plt.show()
```

Observation(s):
- Similar to the contrast Average image above, the face and upper body regions appear red which is reasonably expected since any deepfake video manipulations made would generally focus on these regions too.
